# Crypto Sentiment vs Trader Performance Analysis

This notebook demonstrates a complete quantitative workflow linking market sentiment to trading performance.

```python
from pathlib import Path
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.data_loader import load_sentiment_data, load_trader_data
from src.preprocessing import clean_sentiment_data, clean_trader_data, merge_sentiment_and_trades
from src.sentiment_analysis import sentiment_distribution, fear_vs_greed_frequency
from src.trader_analysis import sentiment_vs_performance, trader_performance_summary, risk_exposure_by_account

sns.set_theme(style='whitegrid')
```

## 1) Load and Validate Datasets

```python
sentiment_raw = load_sentiment_data()
trader_raw = load_trader_data()

print('Sentiment rows:', len(sentiment_raw))
print('Trader rows:', len(trader_raw))
sentiment_raw.head()
```

## 2) Preprocessing

```python
sentiment = clean_sentiment_data(sentiment_raw)
trader = clean_trader_data(trader_raw)
trades = merge_sentiment_and_trades(sentiment, trader)

print('Cleaned sentiment shape:', sentiment.shape)
print('Cleaned trader shape:', trader.shape)
print('Merged shape:', trades.shape)
```

## 3) Market Sentiment EDA

```python
dist = sentiment_distribution(sentiment)
bucket = fear_vs_greed_frequency(sentiment)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.barplot(data=dist, x='sentiment_classification', y='count', ax=axes[0], palette='viridis')
axes[0].set_title('Sentiment Distribution')
axes[0].tick_params(axis='x', rotation=25)

axes[1].pie(bucket['count'], labels=bucket['sentiment_bucket'], autopct='%1.1f%%', startangle=120)
axes[1].set_title('Fear vs Greed Frequency')
plt.tight_layout()
plt.show()
```

## 4) Sentiment vs Performance

```python
sentiment_perf = sentiment_vs_performance(trades)
sentiment_perf
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=trades, x='sentiment_classification', y='closed_pnl', palette='Set2')
plt.title('PnL by Sentiment Regime')
plt.xticks(rotation=25)
plt.show()
```

## 5) Account-Level Analysis

```python
trader_summary = trader_performance_summary(trades)
risk = risk_exposure_by_account(trades)

display(trader_summary.head(10))
display(risk.head(10))
```

## 6) Conclusions

- Compare average and median PnL across sentiment classes.
- Evaluate loss rate and leverage shifts in Fear/Greed phases.
- Profile top accounts by profitability, consistency, and risk.
